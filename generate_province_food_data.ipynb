```python
import json
import os
from random import random

import pandas as pd
```

```python
with open(os.path.join('assets', 'mock.json')) as f:
    json_string = ''.join(f.read())
```

```python
gjson = json.loads(json_string)
gjson
```

```python
gjson.keys()
```

```python
province_ids = [province['properties']['ID'] for province in gjson['features']]
print(max(province_ids), min(province_ids))
```

```python
food = pd.read_csv(os.path.join('data', 'food_price.csv'))
food.head()
```

```python
data = pd.DataFrame(province_ids * len(food), columns=['ID'])
data.head()
```

```python
labels = []
for i in range(len(food)):
    label = str(int(food.iloc[i]['Tahun'])) + '-Q' + str(int(food.iloc[i]['Quarter']))
    labels += [label] * len(province_ids)
```

```python
data['label'] = labels
data.head()
```

```python
for col in food.columns[2:]:
    new_col = []
    for i in range(len(food)):
        original_data = food.iloc[i][col]
        for j in range(len(province_ids)):
            modifier = random() * 5000
            if random() > 0.5:
                new_data = original_data + modifier
            else:
                new_data = original_data - modifier
            new_col.append(new_data)
    data[col] = new_col
```

```python
data
```

```python
json_dict = {}
for label in pd.unique(data['label']):
    period_data = data.loc[data['label'] == label].reset_index(drop=True)
    
    period_dict = {}
    for i in range(len(period_data)):
        row = period_data.iloc[i]
        row_id = str(int(row['ID']))
        id_dict = {}
        for col in period_data.columns[2:]:
            id_dict[col] = float(row[col])
        period_dict[row_id] = id_dict
    
    json_dict[label] = period_dict
```

```python
json.dump(json_dict, open(os.path.join('data', 'food_per_province.json'), 'w'))
```
