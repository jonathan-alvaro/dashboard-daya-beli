```python
def calculate_index(row):
    weights = pd.read_csv(os.path.join('input', 'food_weight.csv'))
    
    Pn = 0
    Po = 0
    for food in pd.unique(weights['food']):
        food_weight = weights.loc[weights['food'] == food]['weight'].values[0]
        Pn += row[food + '_x'] * food_weight
        Po += row[food + '_y'] * food_weight
    
    index = Pn / Po * 100
    return index - 100
```

```python
from functools import reduce

if len(dfs_to_merge) > 10:
    assert False

merged_food_prices = reduce(
    pd.merge, dfs_to_merge
)

merged_food_prices = merged_food_prices.loc[
    ~(
        (merged_food_prices['Tahun'] == 2019)
        & (merged_food_prices['Bulan'] >= 10)
    )
]

merged_food_prices['Prev Bulan'] = merged_food_prices.apply(
    lambda row: row['Bulan'] - 1 if row['Bulan'] > 1 else 12, axis=1
)

merged_food_prices['Prev Tahun'] = merged_food_prices.apply(
    lambda row: row['Tahun'] if row['Prev Bulan'] != 12 else row['Tahun'] - 1,
    axis=1
)

merged_food_prices = pd.merge(
    merged_food_prices, merged_food_prices,
    left_on=['Prev Tahun', 'Prev Bulan', 'Provinsi'],
    right_on=['Tahun', 'Bulan', 'Provinsi']
)

merged_food_prices = merged_food_prices.drop(
    ['Tahun_y', 'Prev Tahun_x', 'Prev Tahun_y', 'Prev Bulan_x', 'Prev Bulan_y'], axis=1
)

merged_food_prices['index'] = merged_food_prices.apply(calculate_index, axis=1)

weights = pd.read_csv(os.path.join('input', 'food_weight.csv'))
food_names = pd.unique(weights['food'])
cols_to_drop = [col + '_y' for col in food_names]

merged_food_prices = merged_food_prices.drop(cols_to_drop, axis=1)

merged_food_prices.columns = [
    col[:-2] if col[-2:] == '_x' else col for col in merged_food_prices.columns
]

merged_food_prices.head()
```

```python
province_mapping = {
    'NUSATENGGARA BARAT': 'NUSA TENGGARA BARAT',
    'DAERAH ISTIMEWA YOGYAKARTA': 'DI YOGYAKARTA',
    'BANGKA BELITUNG': 'KEPULAUAN BANGKA BELITUNG',
    'DI. ACEH': 'ACEH'
}
```

```python
with open(os.path.join('input', 'indonesia-prov.json')) as f:
    province_data = json.load(f)['features']

province_ids = {}
for province in province_data:
    properties = province['properties']
    
    if properties['Propinsi'] in province_mapping:
        province_key = province_mapping[properties['Propinsi']]
    else:
        province_key = properties['Propinsi']
        
    province_ids[province_key] = properties['ID']

province_ids['SEMUA PROVINSI'] = -1
province_ids
```

```python
merged_food_prices['Provinsi'] = merged_food_prices['Provinsi'].replace(province_ids)
```

```python
merged_food_prices.to_csv(
    os.path.join('output', 'food_data.csv'),
    index=False
)
```

```python
month_dict = {
    '1': 'JAN',
    '2': 'FEB',
    '3': 'MAR',
    '4': 'APR',
    '5': 'MEI',
    '6': 'JUN',
    '7': 'JUL',
    '8': 'AUG',
    '9': 'SEP',
    '10': 'OKT',
    '11': 'NOV',
    '12': 'DES'
}
```

```python
json_data = {}
merged_food_prices['Bulan'] = merged_food_prices['Bulan'].astype(str).replace(month_dict)

merged_food_prices['label'] = (
    merged_food_prices['Bulan'].apply(str) + '-'
    + merged_food_prices['Tahun'].apply(str)
)

for period in pd.unique(merged_food_prices['label']):
    period_data = merged_food_prices.loc[
        merged_food_prices['label'] == period
    ].reset_index(drop=True)
    
    period_dict = {}
    
    for location in pd.unique(merged_food_prices['Provinsi']):
        location_data = period_data.loc[
            period_data['Provinsi'] == location
        ]
        location_dict = {}
        
        for col in location_data.columns:
            if location_data[col].dtype != 'float64':
                continue
            location_dict[col] = float(location_data[col].values[0])
        
        period_dict[location] = location_dict
    
    json_data[period] = period_dict
    
with open(os.path.join('output', 'data.json'), 'w') as f:
    json.dump(json_data, f)
```
