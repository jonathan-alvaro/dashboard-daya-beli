```python
bawang_merah = pd.read_excel(
    os.path.join('input', '2016', 'Bawang Merah 2016-2017.xlsx'),
    skiprows=6,
    index_col=1
)
bawang_merah = bawang_merah.drop(['No.'], axis=1)

bawang_merah = bawang_merah.stack().reset_index()
bawang_merah.columns = ['Provinsi', 'Tanggal', 'Harga']

bawang_merah = bawang_merah.loc[
    bawang_merah['Harga'] != '-'
].reset_index(drop=True)
bawang_merah['Harga'] = bawang_merah['Harga'].astype(float)

bawang_merah['Bulan'] = pd.DatetimeIndex(
    bawang_merah['Tanggal']
).month
bawang_merah['Tahun'] = pd.DatetimeIndex(
    bawang_merah['Tanggal']
).year
bawang_merah['Tanggal'] = pd.DatetimeIndex(
    bawang_merah['Tanggal']
).day

bawang_merah = bawang_merah.sort_values(
    by=['Tahun', 'Bulan', 'Tanggal', 'Provinsi']
)
bawang_merah = bawang_merah[[
    'Tahun', 'Bulan', 'Tanggal', 'Provinsi', 'Harga'
]]
bawang_merah = bawang_merah.rename(columns={'Harga': 'Bawang Merah'})
bawang_merah = bawang_merah.reset_index(drop=True)

bawang_merah['Quarter'] = (bawang_merah['Bulan'] - 1) // 3 + 1
bawang_merah = bawang_merah.drop('Bulan', axis=1)

bawang_merah = bawang_merah.groupby(
    ['Tahun', 'Quarter', 'Provinsi']
).mean().reset_index()
bawang_merah = bawang_merah.drop('Tanggal', axis=1)

dfs_to_merge.append(bawang_merah)

bawang_merah.head()
```

```python
from functools import reduce

merged_food_prices = reduce(
    pd.merge, dfs_to_merge
)

merged_food_prices['index'] = 0

food_index_weights = pd.read_csv(
    os.path.join('input', 'food_weight.csv')
)

for col in merged_food_prices.columns[:-1]:
    if merged_food_prices[col].dtype != 'float64':
        continue
    
    weight = food_index_weights.loc[
        food_index_weights['food'] == col
    ]['weight'].values
    
    merged_food_prices['index'] = (
        merged_food_prices['index']
        + merged_food_prices[col] * weight / 100
    )

merged_food_prices.head()
```

```python
province_mapping = {
    'NUSATENGGARA BARAT': 'NUSA TENGGARA BARAT',
    'DAERAH ISTIMEWA YOGYAKARTA': 'DI YOGYAKARTA',
    'BANGKA BELITUNG': 'KEPULAUAN BANGKA BELITUNG',
    'DI. ACEH': 'ACEH'
}
```

```python
with open(os.path.join('input', 'indonesia-prov.json')) as f:
    province_data = json.load(f)['features']

province_ids = {}
for province in province_data:
    properties = province['properties']
    
    if properties['Propinsi'] in province_mapping:
        province_key = province_mapping[properties['Propinsi']]
    else:
        province_key = properties['Propinsi']
        
    province_ids[province_key] = properties['ID']

province_ids['SEMUA PROVINSI'] = -1
province_ids
```

```python
merged_food_prices['Provinsi'] = merged_food_prices['Provinsi'].replace(province_ids)
```

```python
merged_food_prices.to_csv(
    os.path.join('output', 'aggregate_data.csv'),
    index=False
)
```

```python
merged_food_prices = pd.read_csv(
    os.path.join('output', 'aggregate_data.csv')
)
merged_food_prices.head()
```

```python
json_data = {}
merged_food_prices['label'] = (
    merged_food_prices['Tahun'].apply(str) + '-Q'
    + merged_food_prices['Quarter'].apply(str)
)

for period in pd.unique(merged_food_prices['label']):
    period_data = merged_food_prices.loc[
        merged_food_prices['label'] == period
    ].reset_index(drop=True)
    
    period_dict = {}
    
    for location in pd.unique(merged_food_prices['Provinsi']):
        location_data = period_data.loc[
            period_data['Provinsi'] == location
        ]
        location_dict = {}
        
        for col in location_data.columns:
            if location_data[col].dtype != 'float64':
                continue
            location_dict[col] = float(location_data[col].values[0])
        
        period_dict[location] = location_dict
    
    json_data[period] = period_dict

with open(os.path.join('output', 'food.json'), 'w') as f:
    json.dump(json_data, f)
```
